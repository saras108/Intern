# Basics

### Create arrays: x, y  and z where:
- x contains 12 evenly spaced numbers between 1 and 20 
- y contains the values of x multiplied by 2 ie. 2*x 
- z contains the squares of elements of x ie x**2


```python
x = np.linspace(1, 20, num=12, retstep=True)[0]

print('x = ' ,x)
```

```python
y =np.array( [2*x  for x in range(1,13)])

print('y = ' ,y)
```

```python
z = np.array([x**2  for x in range(1,13)])

print('z = ' ,z)
```

#### a. Plot y vs x, set the label of x-axis as ‘x’, the label of y-axis as ‘y’ and the title of the plot as ‘y = 2x’. Increase the x axis range to (0, 25) and y axis range to (0, 450)

```python
# Plot y vs x, set the label of x-axis as ‘x’, the label of y-axis as ‘y’ and the title of the plot as ‘y = 2x’

plt.plot(x, y)

plt.title("y =2x")
plt.xlabel("x")
plt.ylabel("y")

plt.show()
```

```python
# Increase the x axis range to (0, 25) and y axis range to (0, 450)

plt.plot(x, y)

plt.title("y =2x")
plt.xlabel("x")
plt.ylabel("y")

plt.xlim([0, 25])
plt.ylim([0, 450])


plt.show()
```

#### b.Create a new plot and plot y vs x and z vs x in the same plot. Differentiate between the two  functions using legend.


```python
plt.plot(x, y, color='green', marker='o', linestyle='dashed',  linewidth=2, markersize=5 , label = "x vs y")

plt.plot(z, x, color='red', marker='^', linestyle='dashed',  linewidth=2, markersize=5, label = "z vs x")

plt.legend()
plt.show()
```

#### c. Create a figure containing two subplots. Set the size of the figure to be 12X6. In the first subplot, plot y vs x and in the second subplot, plot z vs y. Set the title and axis labels accordingly.


```python
fig, axs = plt.subplots(1, 2)
fig.suptitle('figure containing two subplots')


axs[0].plot(y, x  , 'tab:green' , marker='^', linestyle='dashed',  linewidth=2, markersize=5, label = "y vs x")
axs[0].set_title('y vs x')

axs[1].plot(z, y  , 'tab:orange' , marker='^', linestyle='dashed',  linewidth=2, markersize=5, label = "z vs y")
axs[1].set_title('z vs y')


plt.legend()
plt.show()

```

#### d. Finally, redo all the above plots using seaborn which is a higher level plotting library. It creates more aesthetically pleasing charts than matplotlib.

```python
import seaborn as sns

sns.set_theme()

x = np.linspace(1, 20, num=12, retstep=True)[0]
y =np.array( [2*x  for x in range(1,13)])
z = np.array([x**2  for x in range(1,13)])

data = np.array([x,y,z])

df = pd.DataFrame(data=data , index=["X", "Y", "Z"]).T
```

```python
# a. Plot y vs x, set the label of x-axis as ‘x’, the label of y-axis as ‘y’ and the title of the plot as ‘y = 2x’.

sns.pointplot(data= df, x="X", y="Y").set_title('X vs Y')
```

```python
#  Increase the x axis range to (0, 25) and y axis range to (0, 450)

sns.pointplot(data= df, x="X", y="Y").set(xlim = (0,25) , ylim = (0,450))
```

```python
  # b.Create a new plot and plot y vs x and z vs x in the same plot. Differentiate between the two  functions using legend.

sns.lineplot(data= df, x="Y", y="X" , label = "y vs x")
sns.lineplot(data= df, x="Z", y="X", label = "z vs x")
```

```python
#  c. Create a figure containing two subplots. Set the size of the figure to be 12X6. In the first subplot, plot y vs x and in the second subplot, plot z vs y. Set the title and axis labels accordingly.

fig, axs = plt.subplots(ncols=2 , figsize=(12,6))
sns.lineplot(x='Y', y='X', data=df, ax=axs[0]).set_title('y vs x')
fig.suptitle('2 subplots using seaborn')
sns.lineplot(x='Z', y='X', data=df ,ax=axs[1]).set_title('z vs x')
```

####  b. Download the dataset (PIMA Diabetes Dataset)



```python
dataset = pd.read_csv('pima_diabetes.csv')

dataset.head()
```

- Plot a bar plot of the ‘outcome’ variable. Is the dataset balanced or unbalanced? Try to explore the consequences of having such a dataset.

```python
sns.countplot(x='Outcome', data=dataset)

# From the digram we can say that the data is not balanced as the number of zeros are only around 250 but the the number of ones are 500. These trainng set will be biased.
```


 
- Create a figure containing 8 subplots arranged in (2, 4) grid. Plot the histograms of the 8 features (all the columns except the outcome) present in the dataset. Identify the features having normal distribution and skewed distributions.

```python
fig, axs = plt.subplots(ncols=4 , nrows=2, figsize=(18,9))

colm = list(dataset.columns.values)

k = 0
for i in range(0,2):
  for j in range(0,4):
    # print(i , j)
    sns.histplot(dataset, x=colm[k], ax=axs[i][j])
    k = k +1
    
#Normal Distribution: [Glucose , Blood Pressure , Skin Thickness, BMI, DiabetesPedigreeFunction]
# Positive Skewed: [Pregnancies, Insulin, Age]
```



- Create a figure containing 8 subplots arranged in (4, 2) grid. Plot the boxplot for each of the features in each of the subplots. Outliers are a set of observations that vary significantly from the other observations. Which feature do you think has the most outliers? And Why?

```python
fig, axs = plt.subplots(ncols=2 , nrows=4, figsize=(14,17))

colm = list(dataset.columns.values)

k = 0

for i in range(0,4):
  for j in range(0,2):
    # print(i , j)
    sns.boxplot(x=dataset[colm[k]] , ax=axs[i][j])
    k = k +1
    
#     DiabetesPedigreeFunction and Insulin have the most outliers as in the graph below we can see the most outliers outside the box range in those feature
```
